```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic= sns.load_dataset('titanic')
titanic.head()
```

```python
bp = titanic['fare'].fillna(0).cumsum()
bp.plot()
plt.title('Cumulative Sum of Titanic Fare')
plt.xlabel('Index')
plt.ylabel('Cumulative Fare')
plt.show()
```

```python
df_random = pd.DataFrame(np.random.randn(500, 4), index=bp.index[:500],
columns=list("WXYZ"))

df_random = df_random.cumsum()

plt.figure(figsize=(10,6))
df_random.plot()
plt.title('Cumulative Sum of Random Data')
plt.xlabel('Index (Cumulative Fare)')
plt.ylabel('Cumulative Values')
plt.show()          
```

```python
plt.figure(figsize=(10,6))
sns.histplot(titanic['age'].dropna(), kde=True, bins=10)
plt.title('Distribution of Age')
plt.xlabel('Age1')
plt.ylabel('Frequency')
plt.show()           
```
