```python
df ['Volume'] = np.log (df ['Volume']+1)
```

```python
sns.histplot(df['Volume'],kde=True)
plt.title('Volume and skewness')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['High'], label='High', marker='4')
plt.plot(df['Date'], df['Low'], label='Low', marker='4')
plt.title('High and Low Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
# plt.legend()
# plt.grid(True)
# plt.xticks(rotation=45)
# plt.tight_layout()
plt.show()
```

```python
df = df.sort_values ('Date')

correlation = df ['Close'].corr(df['Volume'])
print (f"correlation between close and volume:{correlation:.4f}")
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x='Volume', y='Close', data=df, scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title('Correlation between Volume and Close Price')
plt.xlabel('Volume')
plt.ylabel('Close Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
corr_matrix = df [['Volume', 'Close']].corr()
plt.figure (figsize = (8,5))
sns.heatmap (corr_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title ('correlation matrix')
```
